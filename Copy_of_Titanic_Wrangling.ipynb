```python
# Import the titanic dataset here
import numpy as np
import pandas as pd
data_dir = "https://dlsun.github.io/pods/data/"
df_titanic = pd.read_csv(data_dir + "titanic.csv")
```

```python
#Checking to see if dataset loaded in properly
df_titanic.head()
```

## 1. Filter the data to include passengers only. Calculate the joint distribution (cross-tab) between a passenger's class and where they embarked.

```python
# Assigning passenger type based on class variable
df_titanic["type"] = df_titanic["class"].map({
    "1st": "passenger",
    "2nd": "passenger",
    "3rd": "passenger",
})

df_titanic
```

```python
#Filtering dataset to show only passengers
df_titanic[(df_titanic["type"] == "passenger")]
```

```python
# Creating a new dataframe with the updated passenger filter
df_passenger = df_titanic[(df_titanic["type"] == "passenger")]
df_passenger
```

```python
# Creating a joint distribution (cross-tab) between class and embarked for passengers
pd.crosstab(df_passenger["class"], df_passenger["embarked"])
```

```python
# Getting probabilities for this joint distribution
pd.crosstab(df_passenger["class"], df_passenger["embarked"], normalize=True)
```

## 2. Using the joint distribution that calculated above, calculate the following:

* the conditional distribution of their class given where they embarked
* the conditional distribution of where they embarked given their class

Use the conditional distributions that you calculate to answer the following quesitons:

* What proportion of 3rd class passengers embarked at Southampton?
* What proportion of Southampton passengers were in 3rd class?

```python
# The conditional distrubution of their class given where they embarked
pd.crosstab(df_passenger["embarked"], df_passenger["class"], normalize="index")
```

```python
# The conditional distribution of where they embarked given their class
pd.crosstab(df_passenger["class"], df_passenger["embarked"], normalize="index")
```

```python
#conditional distribution of class given embarkment location
df_passenger.groupby("embarked")["class"].value_counts(normalize = True)
```
